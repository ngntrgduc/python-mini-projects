```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
def one_year(x):
    return [x**i for i in range(1, 366)]

# 1% better everyday
plt.plot(one_year(1.01))
# 1% worse everyday
plt.plot(one_year(0.9), 'r')
plt.legend([
    '1% better everyday',
    '1% worse everyday'
])
plt.title('The Power of Tiny Gains')
plt.xticks([])
plt.yticks([])
sns.despine()
plt.xlim(left=1)
plt.show()
```
